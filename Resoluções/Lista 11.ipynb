```python
import pandas as pd
import numpy as np
from sklearn import metrics
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
import xgboost as xgb
from xgboost import XGBClassifier
import lightgbm as lgb
from catboost import CatBoostClassifier, Pool
from sklearn.linear_model import Perceptron
from sklearn.neural_network import MLPClassifier
```

```python
df = pd.read_csv('/content/drive/MyDrive/UFC/2021.1/ML/dataset/diabetes.csv')
```

```python
X = df.drop(["Outcome"], axis=1)
y = df["Outcome"]
X_treino, X_teste, y_treino, y_teste = train_test_split(X, y, test_size=0.2, random_state=10)
```

Random Forest

```python
model_rf = RandomForestClassifier(n_estimators=15,max_depth=15, random_state=0, n_jobs=-1)
model_rf.fit(X_treino, y_treino);
y_pred = model_rf.predict(X_teste)
metrics.f1_score(y_teste,y_pred, average="weighted")
```

Gradient Boosting 

```python
model_gb = GradientBoostingClassifier(n_estimators=100, learning_rate=1.0,max_depth=1
                                      , random_state=0).fit(X_treino, y_treino)
y_pred = model_gb.predict(X_teste)
metrics.f1_score(y_teste,y_pred,average="weighted")
```

XGBoost

```python
model_xg = xgb.XGBClassifier(max_depth=5, n_jobs=-1)
model_xg.fit(X_treino, y_treino)
y_pred = model_xg.predict(X_teste)
metrics.f1_score(y_teste,y_pred,average="weighted")
```

LightGBM 

```python
model_gbm = lgb.LGBMClassifier(max_depth=15, max_bin=200)
model_gbm.fit(X_treino, y_treino);
y_pred = model_gbm.predict(X_teste)
metrics.f1_score(y_teste,y_pred,average="weighted")
```

CatBoost

```python
model_cat = CatBoostClassifier(depth=3);
model_cat.fit(X_treino,y_treino);
y_pred = model_cat.predict(X_teste);
```

```python
metrics.f1_score(y_teste,y_pred,average="weighted")
```

Perceptron

```python
model_p = Perceptron(penalty='elasticnet', class_weight='balanced', alpha=1e-6)
model_p.fit(X_treino, y_treino)
y_pred = model_p.predict(X_teste)
metrics.f1_score(y_teste,y_pred, average="weighted")
```

Multi-layer Perceptron

```python
model_mlp = MLPClassifier(random_state=1, beta_1=0.8, beta_2=0.95).fit(X_treino, y_treino)
y_pred = model_mlp.predict(X_teste)
metrics.f1_score(y_teste,y_pred, average="weighted")
```
