```python
import numpy as np
import pandas as pd
```

Implementação KNN

```python
def euclidean_distances(X, X_row):
    X_ = (X - X_row) ** 2
    return np.sum(X_, axis=1) ** 0.5
```

```python
def get_neighbors(X_train, test_row, k):
    distances = euclidean_distances(X_train, test_row)
    idx_sort = np.argsort(distances)
    return idx_sort[:k]
```

```python
def predict_regression(X, y, test_row, k):
    idx_sort = get_neighbors(X, test_row, k)
    output_values = y[idx_sort]
    #counts = np.unique(output_values, return_counts=True)
    #idx_max = np.argmax(counts[1])
    #prediction = counts[0][idx_max]
    predict_value = np.round(np.sum(output_values)/k, 1)
    print('prediction:{}'.format(predict_value))
    return predict_value
```

Criação do dataset

```python
df2 = pd.DataFrame(
{
         "Age": np.array([25,35,45,20,35,52,23,40,60,48,33]),
        "Loan": np.array([40000,60000,80000,20000,120000,18000,95000,62000,100000,220000,150000]),
         "HPI": np.array([135,256,231,267,139,150,127,216,139,250,264]),
        }
)
```

```python
X_train = df2.drop(["HPI"], axis=1)
y_train = df2["HPI"]
```

```python
pred = predict_regression(X_train, y_train, np.array([42,142000]),3)
```

```python
pred = predict_regression(X_train, y_train, np.array([30,50000]),3)
```

```python
#Cáculo dos minimos
min_age = np.min(np.array(X_train["Age"]))
min_loan = np.min(np.array(X_train["Loan"]))
#Cálculo dos máximos 
max_age = np.max(np.array(X_train["Age"]))
max_loan = np.max(np.array(X_train["Loan"]))
#Criação das matrizes com os minimos e os máximos
len_matriz = np.shape(X_train)[0]
X_min = np.full((len_matriz,2), min_age,min_loan)
X_max = np.full((len_matriz,2), max_age, max_loan)
#Criação da matriz normalizada
X_i = (X_train - X_min)/(X_max - X_min)
```

```python
age_test_norm = (42 - min_age)/(max_age - min_age)
loan_test_norm = (50000 - min_loan)/(max_loan - min_loan)

pred = predict_regression(X_i, y_train, np.array([age_test_norm,loan_test_norm]),5)
```
