```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import datasets
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
from sklearn.preprocessing import OrdinalEncoder
from sklearn import tree
from sklearn import svm
from sklearn import linear_model
from sklearn.metrics import classification_report
from sklearn.preprocessing import StandardScaler
```

Carregando o dataset

```python
df = pd.read_csv('/content/drive/MyDrive/UFC/2021.1/ML/dataset/Star3642_balanced.csv')
```

Divisão do dataset entre treino e teste

```python
#Exclusão da coluna SpType por ser uma coluna de strings
X = df.drop(["TargetClass", "SpType"], axis=1)
y = df["TargetClass"]
X_treino, X_teste, y_treino, y_teste = train_test_split(X, y, test_size=0.7, random_state=42, stratify=y)
```

Criação do modelo

```python
#Knn
model_knn = KNeighborsClassifier(n_neighbors=5, weights='distance')
model_knn.fit(X_treino,y_treino);

#Tree
model_tree = tree.DecisionTreeClassifier(random_state=42)
model_tree.fit(X_treino,y_treino);

#Svm
model_svm = svm.SVC(random_state=42)
model_svm.fit(X_treino,y_treino);

#Logistic Regression
model_lr = linear_model.LogisticRegression(random_state=42)
model_lr.fit(X_treino,y_treino);
```

Precisão

```python
target_names = ['Anã', 'Gigante']
```

```python
#knn accuracy
y_pred = model_knn.predict(X_teste)
print(classification_report(y_teste, y_pred, target_names=target_names))
knn_acc = [metrics.accuracy_score(y_teste,y_pred)]
```

```python
#Tree accuracy
y_pred = model_tree.predict(X_teste)
print(classification_report(y_teste, y_pred, target_names=target_names))
tree_acc = [metrics.accuracy_score(y_teste,y_pred)]
```

```python
#Svm accuracy
y_pred = model_svm.predict(X_teste)
print(classification_report(y_teste, y_pred, target_names=target_names))
svm_acc = [metrics.accuracy_score(y_teste,y_pred)]
```

```python
#Logistic Regression accuracy
y_pred = model_lr.predict(X_teste)
print(classification_report(y_teste, y_pred, target_names=target_names))
lr_acc = [metrics.accuracy_score(y_teste,y_pred)]
```

```python
acc_all = np.array([[knn_acc], [tree_acc], [svm_acc], [lr_acc]])
acc = {'Knn':knn_acc, 'Tree':tree_acc, 'SVM':svm_acc, 'LR':lr_acc}
accuracy = pd.DataFrame(data=acc)
accuracy
```

Divisão do dataset

```python
X_2 = df.drop(["TargetClass", "SpType"], axis=1)
y_2 = df["TargetClass"]
X_treino_2, X_teste_2, y_treino_2, y_teste_2 = train_test_split(X_2, y_2, test_size=0.7, random_state=42, stratify=y)

#X_2_treino e X_2_teste Standard
scaler = StandardScaler()
X_treino_std = scaler.fit_transform(X_treino_2)
X_teste_std = scaler.transform(X_teste_2)
```

Criação dos modelos

```python
#knn
model_knn_2 = KNeighborsClassifier(n_neighbors=5, weights='distance')
model_knn_2.fit(X_treino_2,y_treino_2);
#Standard
model_knn_std = KNeighborsClassifier(n_neighbors=5, weights='distance')
model_knn_std.fit(X_treino_std, y_treino_2)

#Tree
model_tree_2 = tree.DecisionTreeClassifier(random_state=42)
model_tree_2.fit(X_treino_2, y_treino_2);
#Standard
model_tree_std = tree.DecisionTreeClassifier(random_state=42)
model_tree_std.fit(X_treino_std, y_treino_2);

#Svm
model_svm_2 = svm.SVC(random_state=42)
model_svm_2.fit(X_treino_2, y_treino_2);
#Standard
model_svm_std = svm.SVC(random_state=42)
model_svm_std.fit(X_treino_std, y_treino_2)

#Logistic Regression
model_lr_2 = linear_model.LogisticRegression(random_state=42, max_iter=1000)
model_lr_2.fit(X_treino_2, y_treino_2);
#Standard
model_lr_std = linear_model.LogisticRegression(random_state=42)
model_lr_std.fit(X_treino_std, y_treino_2);
```

Precisão

```python
#knn accuracy
y_pred = model_knn_2.predict(X_teste_2)
print(classification_report(y_teste_2, y_pred, target_names=target_names))
knn_acc = [metrics.accuracy_score(y_teste_2,y_pred)]
```

```python
#Tree standard
y_pred = model_tree_std.predict(X_teste_std)
print(classification_report(y_teste_2, y_pred, target_names=target_names))
tree_acc_std = [metrics.accuracy_score(y_teste_2,y_pred)]
```

```python
#SVM accuracy
y_pred = model_svm_2.predict(X_teste_2)
print(classification_report(y_teste_2, y_pred, target_names=target_names))
svm_acc = [metrics.accuracy_score(y_teste_2,y_pred)]
```
