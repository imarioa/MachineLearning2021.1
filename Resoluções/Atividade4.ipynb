```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('/content/drive/MyDrive/UFC/2021.1/ML/dataset/winequality-white.csv', sep=';')
```

```python
def euclidean_distances(X, X_row):
    X_ = (X - X_row) ** 2
    return np.sum(X_, axis=1) ** 0.5
```

```python
def get_neighbors(X_train, test_row, k):
    distances = euclidean_distances(X_train, test_row)
    idx_sort = np.argsort(distances)
    return idx_sort[:k]
```

```python
def predict_classification(X, y, test_row, k):
    idx_sort = get_neighbors(X, test_row, k)
    output_values = y[idx_sort]
    counts = np.unique(output_values, return_counts=True)
    idx_max = np.argmax(counts[1])
    prediction = counts[0][idx_max]
    #print('idx_sort:{}, output_values:{}, prediction:{}'.format(idx_sort, output_values, prediction))
    return prediction
```

```python
df['quality'].value_counts()
```

```python
data_training = df.sample(frac=0.7, random_state=42)
data_test = df.drop(data_training.index)
```

```python
#Treino
x_treino = data_training.drop(["quality"], axis=1)
y_treino = data_training["quality"]
#Teste
x_teste = data_test.drop(["quality"], axis=1)
y_teste = data_test["quality"]
```

```python
def prediction(y_teste, x_teste, x_treino, y_treino, k):
  pred = np.zeros(len(np.array(y_teste)))
  it = 0
  for i in np.array(x_teste):
    pred[it] = predict_classification(np.array(x_treino), np.array(y_treino), i, k)
    it=it+1
  return pred
```

```python
# k = 2
k_2 = prediction(y_teste,x_teste,x_treino,y_treino, 2)
hit = (k_2) == np.array(y_teste)
print("Precisão: ", {np.sum(hit)/y_teste.shape[0] * 100})
```
